#### 1. . Find the most expensive car company name

```python
CarPrice = df['price']
Expensive_car = CarPrice.max()
print(np.unique(df[df['price'] == Expensive_car]['make']))
```

```python
df['make'].value_counts()
```

#### 3. Find each company’s highest price car 

```python
car_company = df.groupby(['make'])['price'].max().reset_index()
car_company
```

#### 4. Sort all cars by Price column

```python
sorting_by_price=df.sort_values(by=['price'],inplace=False)
sorting_by_price
```
