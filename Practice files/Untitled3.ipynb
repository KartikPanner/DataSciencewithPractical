```python
flight_data = pd.read_csv('flight_delays.csv')
```

```python
flight_data
```

```python
plt.figure(figsize=(16,8))
plt.title('Average arrival delays for the airline WM, by month')
sns.barplot(x=flight_data.index, y = flight_data['WN'])
#sns.lineplot(data=df['Despacito'], label='Desp')
#sns.lineplot(data=df['Shape of You'], label='Sh_you')
#plt.xlabel("from the release of the popularity for these songs in the year 17 and 18")
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WM, by month')
#sns.barplot(x=flight_data.index, y = flight_data['WN'])
sns.heatmap(data=flight_data)
plt.ylabel('Arrival delays')
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WM, by month')
#sns.barplot(x=flight_data.index, y = flight_data['WN'])
sns.heatmap(data=flight_data, annot=True)
plt.ylabel('Arrival delays')
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WM, by month')
#sns.barplot(x=flight_data.index, y = flight_data['WN'])
sns.heatmap(data=flight_data, annot=True, cmap='YlGnBu')
plt.ylabel('Arrival delays')
```

```python
flight_data.corr()
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WM, by month')
#sns.barplot(x=flight_data.index, y = flight_data['WN'])
corr = flight_data.corr()
sns.heatmap(data=corr, annot=True, cmap='PRGn')
plt.ylabel('Arrival delays')
```
