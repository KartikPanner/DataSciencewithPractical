```python
dict3={'Name':['Alex','Amy','Allen','Alice','Ayoung'],'id':[1,2,3,4,5],'Subject id':['Eng','Math','Sci.','Stat','Prob']}
df_left=pd.DataFrame(dict3)
df_left
```

```python
dict4={'Name':['Billy','Brain','Bran','Bryce','Betty'],
       'id':[1,2,3,4,5],'Subject id':['Math','Sci.','Eng','Stat','Prob']}
df_right=pd.DataFrame(dict4)
df_right
```

```python
pd.merge(df_left, df_right, on='id')
```

```python
pd.merge(df_left, df_right, on=['id','Subject id'])
```

```python
df_left.merge(df_right, on='Subject id',how = 'left')
```

```python
pd.concat([df_left, df_right])
```

```python
pd.concat([df_left, df_right], axis=1)
```

```python
pd.concat([df_left, df_right], ignore_index= True)
```
